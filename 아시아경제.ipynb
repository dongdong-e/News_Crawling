## **아시아경제**

```python
# html 뽑아내기
from bs4 import BeautifulSoup
import urllib.request
import urllib.parse
import re

source = []

for i in range(1, 10):
    i = str(i)
    j = i.zfill(2)
    with urllib.request.urlopen(f'https://news.naver.com/main/list.nhn?mode=LPOD&mid=sec&oid=277&listType=paper&date=201907{j}') as response:
        html = response.read()
        soup = BeautifulSoup(html, 'html.parser')    
        
        for a in soup.find_all('dt'):
            source.append(a)

for i in range(10, 21):
    with urllib.request.urlopen(f'https://news.naver.com/main/list.nhn?mode=LPOD&mid=sec&oid=277&listType=paper&date=201907{j}') as response:
        html = response.read()
        soup = BeautifulSoup(html, 'html.parser')    
        
        for a in soup.find_all('dt'):
            source.append(a)

source_str = ''.join(map(str, source))
source_str = source_str.replace(' ', '')
link_amp = re.findall('https://n.{80}', source_str)
# 중복 링크 제거
link_amp = list(set(link_amp))

link_fin = []

for i in range(0, len(link_amp)):
    link_fin.append(link_amp[i].replace('amp;', ''))

link_fin
```

```python
# 제목 뽑아내기
from bs4 import BeautifulSoup
import urllib.request
import urllib.parse

# web_url에 원하는 웹의 URL을 넣어주시면 됩니다.
title_fin = []

for i in link_fin:
    with urllib.request.urlopen(i) as response:
        print(i)
        html = response.read()
        soup = BeautifulSoup(html, 'html.parser')
        title_fin.append(soup.find(id='articleTitle').get_text())

title_fin
```
